# NYC Diabetes Death Rate




```python
import pandas as pd
import numpy as np
import urllib
import json
```

The following code imports the NYC Deaths dataset and generates it for viewing:

```python
nyc = pd.read_csv('https://raw.githubusercontent.com/Gitprincesskerry/IS362/main/Final%20Project/new-york-city-leading-causes-of-death.csv')
nyc
```

Next, the following code load the information about the dataset about anx allows the viwer to read what the dataset is about.

```python
nycjson = 'https://raw.githubusercontent.com/Gitprincesskerry/IS362/main/Final%20Project/socrata_metadata.json'


nycdata = urllib.request.urlopen(nycjson).read()
nycdata

nyctojson = json.loads(nycdata)
nyctojson
```

Next, we will copy the dataset so that we can manipulate the data.

```python
nyc2 = nyc.copy()
```

Let's take a look at the columns of the datset:

```python
nyc2.columns
```

The dataset column titles has spaces which will not one to "call" the field based on the nme thus the following code will replace the space in any column name with an underscore so that we can manipulate the data without any issues calling the field.

```python
nyc2.columns = nyc2.columns.str.replace(' ', '_')
nyc2
```

Next, let's do a groupby which will give us all the diseases in the datset.

```python
nyc2.groupby('Leading_Cause').count()
```

The following code gives us the years of the information in the dataset.

```python
nyc2.Year.unique()
```

The following code gives us the Sex of the information in the dataset.

```python
nyc2.Sex.unique()
```

The following code gives us the Race of the information in the dataset.

```python
nyc2.Race_Ethnicity.unique()
```

The following code gives us the values of the Death count in the dataset.

```python
nyc2.Deaths.unique()
```

The following code gives us the values of the Death Rates in the dataset.

```python
nyc2.Death_Rate.unique()
```

Since, the values of the Sex field aren't all Male and Female, the following code will replace the values of M with Male and values of F with Female.

```python
nyc2.replace(to_replace="M", value ="Male")
nyc2.replace(to_replace="F", value ="Female")
```

The following code will filter the dataframe to see the results of Women with Diabetes in 2017.

```python
nyc_female_2017 = nyc2[(nyc2["Leading_Cause"]=="Diabetes Mellitus (E10-E14)") & (nyc2["Year"]==2017) & (nyc2["Sex"]=="Female")]
nyc_female_2017
```

The following code will filter the dataframe to see the results of Men with Diabetes in 2017.

```python
nyc_male_2017 = nyc2[(nyc2["Leading_Cause"]=="Diabetes Mellitus (E10-E14)") & (nyc2["Year"]==2017) & (nyc2["Sex"]=="Male")]   
nyc_male_2017
```

Based on the information above women die at a higher death rate than men in the Year 2017 across all races except for Hispanic's.
