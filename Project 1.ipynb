**<center><b>To analyze the sample airline data we must import the pandas and numpy libraries. The following code will import both libaries.<center><b>**


```python
import pandas as pd
import numpy as np
```

**<center><b>Next, the code below will import the csv files using pandas.<center><b>**

```python
airline_info_df = pd.read_csv('Project1.csv', index_col='Airline_Name')
airline_info_df
```

**<center><b>Next, the following code below will create a copy of the data from the csv and add 2 new fields: 1) Total_Flight_Count and 2) Delay_Percentage. The Total_Flight_Count column depicts the total flight count whereas Delay_Percentage depicts the delay percentage for each row .<center><b>**

```python
airline_info_df_2 = airline_info_df.copy()
airline_info_df_2

airline_info_df_2.insert(2,'Total_Flight_Count', airline_info_df_2.sum(axis=1))
airline_info_df_2.insert(3,'Delay_Percentage', (airline_info_df.Delayed_Flight_Count / airline_info_df_2.Total_Flight_Count) * 100)
pd.options.display.float_format = '{:.1f}%'.format
airline_info_df_2
```

**<center><b>The following code saves the previous dataframe into a variable called airline_info_df_3 but drops the On_Time_Flight_Count field.<center><b>**

```python
airline_info_df_3 = airline_info_df_2.drop(['On_Time_Flight_Count'], axis=1)
airline_info_df_3
```

**<center><b>Futhermore, the folowing codes will determine the overall delay and on time performance percenetages consecutively..<center><b>**

```python
airline_info_df_4 = airline_info_df_2.groupby(['Airline_Name'])['On_Time_Flight_Count', 'Delayed_Flight_Count'].sum()
airline_info_df_4

airline_info_df_4.insert(2,'Overall_Delay_Percentage', ((airline_info_df_4.Delayed_Flight_Count) / (airline_info_df_4.Delayed_Flight_Count + airline_info_df_4.On_Time_Flight_Count) * 100))
airline_info_df_4
```

```python
airline_info_df_5 = airline_info_df_2.groupby(['Airline_Name'])['On_Time_Flight_Count', 'Delayed_Flight_Count'].sum()
airline_info_df_5

airline_info_df_5.insert(2,'Overall_On_Time_Percentage', ((airline_info_df_4.On_Time_Flight_Count) / (airline_info_df_4.Delayed_Flight_Count + airline_info_df_4.On_Time_Flight_Count) * 100))
airline_info_df_5
```
