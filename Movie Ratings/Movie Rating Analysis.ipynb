**<center><b>To analyze the data in question we must import the pandas and numpy libraries. The following code will import both libaries.<center><b>**

```python
import pandas as pd
import numpy as np
```

**<center><b>Next, the code below will import the csv file movie_ratings using pandas, convert the data into a datframe, and generate the output.<center><b>**

```python
ratings = pd.read_csv('movie_ratings.csv', index_col='Name')
#print(ratings)
ratings = pd.DataFrame(ratings)
ratings
```

**<center><b>Now, let's use the following code to calculate the Average Rating per user.<center><b>**

```python
ratings['Rating Average per User'] = ratings.mean(axis=1)
ratings
```

**<center><b>We will now use the following code to calculate the Average Movie Rating per Movie.<center><b>**

```python
ratings.loc['Rating Average per Movie'] = ratings.mean(axis=0)
ratings
```

**<center><b>We will now proceed to make a copy of the table above so we can normailze it.<center><b>**

```python
normalization_copy = ratings.copy()
normalization_copy
```

**<center><b>Next, let's replace all null values with zeroes so we can use them in the normalization calculation.<center><b>**

```python
normalization_copy = normalization_copy.replace(np.NaN, 0)
normalization_copy
```

**<center><b>Next, let's utilize the normailzation formula on the previous output.<center><b>**

```python
normalization_copy = (normalization_copy - normalization_copy.min()) / (normalization_copy.max() - normalization_copy.min())
normalization_copy
```

**<center><b>Normalization is good to analyze data with a smaller range from 0 to 1. <br> It makes it easier to analyze the minimum and maximum values. <br> Additionally, normalization can help to spot a prospective outlier as well. <center><b>**

**<center><b>Next, let's make a copy of the information so we can standardize it.<center><b>**

```python
standardization_copy = ratings.copy()
standardization_copy
```

**<center><b>Next, let's replace all null values with zeroes so we can use them in the standardization calculation.<center><b>**

```python
standardization_copy = standardization_copy.replace(np.NaN, 0)
standardization_copy
```

**<center><b>Next, let's utilize the standardization formula on the previous output.<center><b>**

```python
standardization_copy = (standardization_copy - standardization_copy.mean()) / (standardization_copy.std())
standardization_copy
```
