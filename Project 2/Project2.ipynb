**<b><u>Dataset 1 - Religious Data Analysis:</u></b>**
===========================================

**<center><b>To analyze the data in qustion we must import the pandas and numpy libraries. The following code will import both libaries.<center><b>**

```python
import pandas as pd
import numpy as np
```

**<center><b>Next, the code below will import the csv files using pandas, drop the column named Unnamed:0,<br> convert the data into a datframe, and generate the output.<center><b>**

```python
covid = pd.read_csv('relinc.csv', index_col='religion')
#print(covid.columns)
covid.drop('Unnamed: 0', axis=1, inplace=True)
covid = pd.DataFrame(covid)
covid
```

**<center><b>Next, we will rename the first and last column titles so we can understand the variables more clearly.<center><b>**

```python
covid.rename(columns = {'<10k':'Income: <10k','refused':'Unknown - Refused to Answer'}, inplace = True)                               
covid
```

**<center><b>Let's add a column which depicts the total number of individuals in each denomination category.<center><b>**

```python
covid['Total'] = covid.sum(axis=1)
covid
```

**<center><b>Let's do some analaysis on the data. First let's find the total number of individuals in the dataset:<center><b>**

```python
total_sample_count = covid['Total'].sum()
total_sample_count
```

**<center><b>Next, we will create a column which depicts the percent of individuals per denomination:<center><b>**

```python
covid.insert(11,'Religious Population Percent', (covid.Total / total_sample_count) * 100)
pd.options.display.float_format = '{:.1f}%'.format
covid
```

**<center><b>Based on the information above, <br> the majority of people in the data set are Catholic (85.6%).  </center></b>**
